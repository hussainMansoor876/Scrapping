```python
print(allData[0][10]['href'])

obj = {
    'companyName': [],
    'url': [],
    'address': [],
    'tel': [],
    'fax': [],
    'website': [],
    'year': [],
    'ust': [],
    'employee': [],
    'detail': [],
    'txt': [],

}


for arr in allData:
    for arr1 in arr:
        print(arr1['href'])
        data = get(arr1['href'], headers={
                "accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3",
                "accept-encoding": "gzip, deflate, br",
                "accept-language": "en-US,en;q=0.9,eu;q=0.8,ur;q=0.7",
                "cache-control": "max-age=0",
                "cookie": "timezoneoffset=-300; __sc_uid_=134e0c3c08b496c4c5dddac0f597f190; _k_cty_lang=de_DE; ROUTEID=.3; cookie_kompass_allowedCookies=Wysistat%2Cws_cookie_dns%2Cwschkvisit24,c1%2Cs1%2Cns_cookietest%2Cns_session,__atuvc%2Cuid,_jsuid%2C_eventqueue,__utma%2C__utmb%2C__utmc%2C__utmz%2C__utmv,undefined; SnapABugHistory=1#; _ga=GA1.2.2015073784.1570192599; _gid=GA1.2.66095710.1570192599; JSESSIONID=126B8EE30A4605B2735D0FCEEA66019C; SnapABugRef=https%3A%2F%2Fde.kompass.com%2Fd%2Fwesterwaldkreis%2Fde_07_07143%2F%20; SnapABugUserAlias=%23; SnapABugVisit=11#1570192417; kp_uuid=62e0352f-8a5b-4e6e-92a0-f63bf723de65",
                "sec-fetch-mode": "navigate",
                "sec-fetch-site": "none",
                "sec-fetch-user": "?1",
                "upgrade-insecure-requests": "1",
                "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36"
            })
        html_data = BeautifulSoup(data.text, 'html.parser')

#         div1 = html_data.find_all('div', class_="container containerCompany")[0]
        
        try:
            span = html_data.find_all('div', itemprop="address")[0]
            span = span.find_all('span', class_="spRight")[0]
            address = span.text.strip().split('\n\r\n')
            address = address[0].strip()+'\n'+address[1].strip()
        except:
            address = ''
            
        try:
            company = html_data.findChild('div', class_="companyCol1 blockNameCompany").findChild('h1').text.strip().split('\n')
            companyName = ''
            for c in company:
                companyName = companyName + " " + c.strip()
            companyName = companyName.strip()
        except:
            companyName = ''
            
        try:
            fax = html_data.find('span', class_="faxNumber")
            fax = fax.text.strip()
            tel = fax.split(' ')[0]
        except:
            fax = ''
            tel = ''
            
        try:
            website = html_data.find('a', id="webSite_presentation_0")['href']
        except:
            website = ''
            
        
        try:
            ust = html_data.find('feature:companyinformationaccess')
            ust = ust.text.strip()
        except:
            ust = ''
            
        try:
            table = html_data.find_all('table')[0].find_all('td')
            employee =  table[5].text.strip()
        except:
            employee = ''
            
        try:
            table = html_data.find_all('table')[0].find_all('td')
            year = table[1].text
        except:
            year = ''
            
        try:
            detail = html_data.find('div', itemprop="description").text
        except:
            detail = ''
            
        try:
            blockInterieur = html_data.find_all('div', class_="executiveBlock")
            txt = ''
            for i in blockInterieur:
                p = i.find_all('p')
                for t in p:
                    txt = txt +"\n "+ t.text.strip()
        except:
            txt = ''
            
        obj['companyName'].append(companyName)
        obj['url'].append(arr1['href'])
        obj['address'].append(address)
        obj['tel'].append(tel)
        obj['fax'].append(fax)
        obj['website'].append(website)
        obj['year'].append(year)
        obj['ust'].append(ust)
        obj['employee'].append(employee)
        obj['detail'].append(detail)
        obj['txt'].append(txt)
        
```
